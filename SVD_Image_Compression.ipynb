```python
import matplotlib.pyplot as plt
import numpy as np

#Change "vermeer.png" to a png image file of your choice below.
#Make sure the file is uploaded into your project 3 folder.

cimg=plt.imread('saturn.png')


#This code converts it to a grey-scale image. Don't touch this.


def to_greyscale(cimg):
    img = np.ones((cimg.shape[0], cimg.shape[1]))
    if np.isscalar(cimg[0][0]):
        img = cimg
    else:
        for i in range(cimg.shape[0]):
            for j in range(cimg.shape[1]):
                img[i,j] = 0.2126*cimg[i][j][0] + 0.7152*cimg[i][j][1] + 0.0722*cimg[i][j][2]
    return img

#This code saves your image as a grey-scale matrix named M.

grey_img = to_greyscale(cimg)

plt.imshow(grey_img, cmap = 'gray')
plt.show()
```

```python
import math
upper = math.floor(min(cimg.shape[0], cimg.shape[1])/10)

incrs = list(10*np.array(range(upper))) 
```

```python
U, D, VT = np.linalg.svd(grey_img)

for r in incrs:
    reduced_img = U[:,:r]@np.diag(D[:r])@VT[:r,:]

    plt.imshow(reduced_img, cmap = 'gray')
    plt.title(f"With {r} singular values")
    plt.show()
```
